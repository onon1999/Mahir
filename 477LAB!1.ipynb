```python
import numpy as np

# Generate 50 random numbers
np.random.seed(0)  # For reproducibility
random_numbers = np.random.rand(50)

# Display the generated random numbers
print("Generated random numbers:")
print(random_numbers)

# Define smoothing function using bin means method
def smooth_data(data, window_size=3):
    smoothed_data = np.zeros_like(data)
    for i in range(len(data)):
        start_index = max(0, i - (window_size - 1) // 2)
        end_index = min(len(data), i + (window_size - 1) // 2 + 1)
        smoothed_data[i] = np.mean(data[start_index:end_index])
        print(f"Step {i+1}: Smoothing {data[start_index:end_index]} to {smoothed_data[i]}")
    return smoothed_data

# Apply smoothing
smoothed_numbers = smooth_data(random_numbers)

# Display the smoothed numbers
print("\nSmoothed numbers:")
print(smoothed_numbers)
```

```python
# Display summary statistics
summary_stats = df.describe()
print(summary_stats)
```

```python
# Show null values
null_values = df.isnull().sum()
print("Null values in each column:")
print(null_values)
```

```python
# Calculate the mean of the "Life expectancy" column
mean_life_expectancy = df['Life expectancy '].mean()

# Fill null values in the "Life expectancy" column with the mean
df['Life expectancy '].fillna(mean_life_expectancy, inplace=True)

# Display the mean of the "Life expectancy" column
print("Mean Life Expectancy:", mean_life_expectancy)

# Display the filled values
print("Filled Life Expectancy Values:")
print(df['Life expectancy '])
```

```python
# Calculate the mode of the "Adult Mortality" column
mode_adult_mortality = df['Adult Mortality'].mode()[0]
# Display the mean of the "Life expectancy" column
print("mode Adult Mortality:", mode_adult_mortality)

# Fill null values in the "Adult Mortality" column with the mode
df['Adult Mortality'].fillna(mode_adult_mortality, inplace=True)
# Display the filled values
print("Adult Mortality Values:")
print(df['Adult Mortality'])

```

```python
# Print the column names in the DataFrame
print(df.columns)
```

```python
# Show the number of missing values in the "BMI " column before dropping
missing_values_before = df[' BMI '].isnull().sum()
print("Number of missing values in the 'BMI' column before dropping:", missing_values_before)

# Drop rows with missing values in the "BMI " column
df.dropna(subset=[' BMI '], inplace=True)

# Show the number of missing values in the "BMI " column after dropping
missing_values_after = df[' BMI '].isnull().sum()
print("Number of missing values in the 'BMI' column after dropping:", missing_values_after)
```

```python
# Replace missing values in the "Schooling" column with forward fill
df['Schooling'].fillna(method='ffill', inplace=True)

# Print the filled "Schooling" column
print("Filled 'Schooling' column:")
print(df['Schooling'])

```
