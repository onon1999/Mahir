```python
import zipfile
import os

# Define the path to the zip file
zip_file_path = '/content/80.zip'

# Verify the file type and size
print(f"File path: {zip_file_path}")
print(f"File size: {os.path.getsize(zip_file_path)} bytes")

# Unzip the file
try:
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall('/content')  # You can specify another directory if needed
    print("Unzipping completed successfully.")
except zipfile.BadZipFile:
    print("Error: The file is not a zip file or it is corrupted.")
```
