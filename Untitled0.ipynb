```python
import os
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Path to the directory containing normalized and resized images
image_directory = '/content/drive/MyDrive/Colab Notebooks/detection/_bee_20230609a/images/'

# List of image filenames without paths
image_filenames = [filename for filename in os.listdir(image_directory) if filename.endswith(".jpg")]

# Normalize pixel values and resize images
target_size = (256, 256)

normalized_resized_images = []
for filename in image_filenames:
    image_path = os.path.join(image_directory, filename)
    image = cv2.imread(image_path)
    image = image.astype(np.float32) / 255.0
    image = cv2.resize(image, target_size)
    normalized_resized_images.append(image)

# Split data into training, validation, and testing sets
train_images, test_images, train_filenames, test_filenames = train_test_split(
    normalized_resized_images, image_filenames, test_size=0.2, random_state=42
)

# Further split validation from training set
train_images, val_images, train_filenames, val_filenames = train_test_split(
    train_images, train_filenames, test_size=0.2, random_state=42
)

# Define the data generator for training
batch_size = 32
train_datagen = ImageDataGenerator(rescale=1.0 / 255)  # Add other augmentations if needed

# Create the generator for training data
train_generator = train_datagen.flow(np.array(train_images), batch_size=batch_size)

# Define the data generator for validation
val_datagen = ImageDataGenerator(rescale=1.0 / 255)

# Create the generator for validation data
val_generator = val_datagen.flow(np.array(val_images), batch_size=batch_size)

# Define the data generator for testing
test_datagen = ImageDataGenerator(rescale=1.0 / 255)

# Create the generator for testing data
test_generator = test_datagen.flow(np.array(test_images), batch_size=batch_size)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# Define your CNN model
def create_cnn_model(input_shape, num_classes):
    model = models.Sequential()

    # Add convolutional and pooling layers
    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
    model.add(layers.MaxPooling2D((2, 2)))

    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))

    model.add(layers.Conv2D(128, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))

    # Flatten the output before the fully connected layers
    model.add(layers.Flatten())

    # Add fully connected layers
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(num_classes, activation='softmax'))  # Output layer

    return model

# Input shape: (height, width, channels)
input_shape = (256, 256, 3)  # Adjust based on your image dimensions
num_classes = 4  # Adjust based on the number of directions you're estimating

# Create the CNN model
model = create_cnn_model(input_shape, num_classes)

# Print the model summary
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',  # Adjust for your task
              metrics=['accuracy'])
```

```python
# Print input shape of the model
print("Model Input Shape:", model.input_shape)

# Print shape of the data generated by the generator
data_batch = next(train_generator)
print("Generated Data Shape:", data_batch[0].shape)  # Assuming data_batch[0] contains the images

# Check if the shapes match
if data_batch[0].shape[1:] != model.input_shape[1:]:
    print("Input shape mismatch. Reshaping data...")
    reshaped_data = data_batch[0].reshape((-1,) + model.input_shape[1:])
    print("Reshaped Data Shape:", reshaped_data.shape)
    # Now use reshaped_data for training
```

```python
epochs = 10  # Adjust based on your experimentation
history = model.fit(train_generator,
                    steps_per_epoch=len(train_images) // batch_size,
                    epochs=epochs,
                    validation_data=val_generator,
                    validation_steps=len(val_images) // batch_size)
```
