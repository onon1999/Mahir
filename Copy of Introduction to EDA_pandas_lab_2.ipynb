# Exercise

Create a class name employee and add the following methods
calcute employee net salary where you take basic salary as input.

use the following equation:
net_salary = 25% of basic_salary

then create a scatter plot for 5 employees with basic salary
basic_salary = [200000,400000,100000,500000,600000]

x axis will be basic_salary
y axis will be net_salary


```python
import matplotlib.pyplot as plt

class Employee:
    def __init__(self, basic_salary):
        self.basic_salary = basic_salary

    def calculate_net_salary(self):
        return 0.25 * self.basic_salary


basic_salaries = [200000,400000,100000,500000,600000]


employees = []
for basic_salary in basic_salaries:
    employee = Employee(basic_salary)
    employees.append(employee)


net_salaries = []
for employee in employees:
    net_salary = employee.calculate_net_salary()
    net_salaries.append(net_salary)

# Create a scatter plot
plt.scatter(basic_salaries, net_salaries, marker='x', color='red')
plt.title('Employee Net Salary vs. Basic Salary')
plt.xlabel('Basic Salary')
plt.ylabel('Net Salary')

# Show the plot
plt.show()
```

# Guideline to Exploratpry Data Analysis

standard guideline for exploratory data analysis (EDA) that you can follow:

- **Understand the Data:** Familiarize yourself with the dataset and the variables it contains. Read any available documentation or data dictionary to gain insights into the meaning and context of the data.

- **Data Cleaning:** Preprocess and clean the data to handle missing values, outliers, and inconsistencies. This step ensures the data is in a usable format for analysis.

- **Descriptive Statistics:** Calculate summary statistics (mean, median, mode, standard deviation, etc.) to gain a high-level understanding of the data distribution and central tendencies.

- **Data Visualization:** Visualize the data using various charts, graphs, and plots to identify patterns, trends, and relationships. Common visualizations include histograms, box plots, scatter plots, bar charts, and line plots.

- **Univariate Analysis:** Analyze individual variables in isolation to understand their distributions, skewness, and potential outliers. Use appropriate visualizations and statistical measures for univariate analysis.

- **Bivariate Analysis:** Explore relationships between pairs of variables to identify correlations, associations, or dependencies. Scatter plots, heatmaps, correlation matrices, and statistical tests (e.g., Pearson correlation coefficient) can be useful for bivariate analysis.

- **Multivariate Analysis:** Investigate interactions and dependencies among multiple variables simultaneously. Techniques such as dimensionality reduction (e.g., PCA), cluster analysis, and parallel coordinates plots can aid in understanding complex relationships.

- **Feature Engineering:** Create new derived features or transform existing features to enhance the predictive power of the data. This step may involve techniques like scaling, binning, one-hot encoding, or creating interaction variables.

- **Data Quality Check:** Verify the quality of the data by checking for data integrity, consistency, and accuracy. Address any issues or discrepancies that may affect the reliability of the analysis.

- **Statistical Testing:** Conduct statistical tests, such as t-tests, chi-square tests, or ANOVA, to evaluate hypotheses, compare groups, or identify significant differences in the data.

- **Documentation:** Document your findings, insights, and any decisions made during the EDA process. Prepare clear and concise summaries, visualizations, and reports that effectively communicate the results of your analysis.



*Remember that EDA is an iterative process, and you may need to revisit previous steps as new insights emerge or data issues are addressed. Additionally, the specific techniques and methods used in EDA may vary depending on the nature of the data and the research question at hand.*

**Data Selection and Indexing:** Pandas allows for easy selection and indexing of data. You can access specific rows, columns, or subsets of data using label-based indexing (loc) or integer-based indexing (iloc). Additionally, you can filter data based on conditions using Boolean indexing.

```python
# Create a sample DataFrame
data = {'Name': ['John', 'Emily', 'Kate', 'Michael', 'Sophia'],
        'Age': [25, 30, 35, 28, 32],
        'City': ['New York', 'London', 'Paris', 'Sydney', 'Tokyo'],
        'Salary': [50000, 60000, 75000, 48000, 70000]}
df = pd.DataFrame(data)

# Select specific columns
print("Name column:")
print(df['Name'])
```

```python
# Select specific rows using loc
print("\nRows with index 1 and 3 using loc:")
print(df.loc[[1, 3]])
```

```python
# Select specific rows and columns using loc
print("\nRows 1 to 3 and columns 'Name' and 'Age' using loc:")
print(df.loc[1:3, ['Name', 'Age']])
```

```python
# Select specific rows using iloc
print("\nRows with index 2 and 4 using iloc:")
print(df.iloc[[2, 4]])
```

```python
# Filter data based on a condition using Boolean indexing
print("\nFiltering rows where Age is greater than 30:")
print(df[df['Age'] > 30])
```

In the code above, we create a sample DataFrame using a dictionary data. The DataFrame consists of columns 'Name', 'Age', 'City', and 'Salary'. We then perform various data selection and indexing operations using Pandas:

Select specific columns: We use square brackets [] to select specific columns. For example, df['Name'] selects the 'Name' column.

Select specific rows using loc: We use the loc indexer to select specific rows based on their labels. For example, df.loc[[1, 3]] selects rows with index 1 and 3.

Select specific rows and columns using loc: We use the loc indexer with row and column labels to select specific rows and columns. For example, df.loc[1:3, ['Name', 'Age']] selects rows 1 to 3 and columns 'Name' and 'Age'.

Select specific rows using iloc: We use the iloc indexer to select specific rows based on their integer positions. For example, df.iloc[[2, 4]] selects rows with integer positions 2 and 4.

Filter data based on a condition using Boolean indexing: We use a conditional statement df['Age'] > 30 inside square brackets to filter rows where the 'Age' column value is greater than 30.

## Data Cleaning and Preprocessing:

Pandas provides functionalities for handling missing data, removing duplicates, and dealing with outliers. You can use methods like **dropna()**, **fillna()**, **duplicated()**, and **drop_duplicates()** to clean and preprocess your data.

```python
# Create a sample DataFrame with missing data and duplicates
data = {'Name': ['John', 'Emily', 'Kate', np.nan, 'Sophia', 'John'],
        'Age': [25, 30, np.nan, 28, 32, 25],
        'City': ['New York', 'London', 'Paris', 'Sydney', 'Tokyo', 'New York']}
df = pd.DataFrame(data)

# Drop rows with missing values
df_dropna = df.dropna()
print("DataFrame after dropping rows with missing values:")
print(df_dropna)
```

```python
# Fill missing values with a specific value
df_fillna = df.fillna(0)
print("\nDataFrame after filling missing values with 0:")
print(df_fillna)
```

```python
# Check for duplicates
print("\nChecking for duplicates:")
print(df.duplicated())

# Drop duplicates
df_drop_duplicates = df.drop_duplicates()
print("\nDataFrame after dropping duplicates:")
print(df_drop_duplicates)
```

In the code above, we create a sample DataFrame using a dictionary data. The DataFrame consists of columns 'Name', 'Age', and 'City'. We then perform various data cleaning and preprocessing operations using Pandas:

Drop rows with missing values: We use the dropna() method to remove rows that contain missing or NaN (Not a Number) values. The resulting DataFrame, df_dropna, will not have any rows with missing values.

Fill missing values with a specific value: We use the fillna() method to replace missing or NaN values with a specified value, which in this case is 0. The resulting DataFrame, df_fillna, will have 0 in place of missing values.

Check for duplicates: We use the duplicated() method to identify duplicate rows in the DataFrame. It returns a Boolean Series indicating whether each row is a duplicate or not.

Drop duplicates: We use the drop_duplicates() method to remove duplicate rows from the DataFrame. The resulting DataFrame, df_drop_duplicates, will not have any duplicate rows.

## Data Manipulation and Transformation

Pandas offers a wide range of functions for manipulating and transforming data. You can perform operations like sorting, merging, joining, grouping, reshaping, and pivoting data. Functions like groupby(), sort_values(), merge(), and pivot_table() are commonly used for data manipulation tasks.

```python
# Create two sample DataFrames
data1 = {'Name': ['John', 'Emily', 'Kate', 'Michael', 'Sophia'],
         'Age': [25, 30, 35, 28, 32],
         'City': ['New York', 'London', 'Paris', 'Sydney', 'Tokyo']}
df1 = pd.DataFrame(data1)

data2 = {'Name': ['John', 'Emily', 'Kate', 'Michael', 'Sophia'],
         'Salary': [50000, 60000, 75000, 48000, 70000]}
df2 = pd.DataFrame(data2)

```

```python
# Sort the DataFrame by a column
df_sorted = df1.sort_values(by='Age')
print("Sorted DataFrame:")
print(df_sorted)

```

```python
# Merge two DataFrames based on a common column
df_merged = pd.merge(df1, df2, on='Name')
print("\nMerged DataFrame:")
print(df_merged)

```

```python
# Perform groupby operation and calculate mean
df_grouped = df1.groupby('City').mean()
print("\nGrouped DataFrame:")
print(df_grouped)

```

```python
# Reshape the DataFrame using pivot_table
df_pivot = pd.pivot_table(df1, values='Age', index='City', columns='Name')
print("\nPivoted DataFrame:")
print(df_pivot)
```

In the code above, we create two sample DataFrames, df1 and df2, using dictionaries data1 and data2. The DataFrames consist of columns 'Name', 'Age', 'City', and 'Salary'. We then perform various data manipulation and transformation operations using Pandas:

Sort the DataFrame: We use the sort_values() method to sort the DataFrame by a specific column. In this example, we sort df1 based on the 'Age' column.

Merge two DataFrames: We use the merge() function to merge two DataFrames based on a common column. In this example, we merge df1 and df2 on the 'Name' column.

Perform groupby operation: We use the groupby() method to group the DataFrame by a specific column. In this example, we group df1 by the 'City' column and calculate the mean of the other numeric columns.

Reshape the DataFrame using pivot_table: We use the pivot_table() function to reshape the DataFrame based on specific columns. In this example, we create a pivoted DataFrame where the values are the 'Age' column, the index is the 'City' column, and the columns are the 'Name' column.

A violin plot is a type of data visualization that combines aspects of a box plot and a kernel density plot. It is used to display the distribution and density of a continuous variable across different categories or groups. The plot gets its name from its resemblance to the shape of a violin or a guitar.

In a violin plot, each category or group is represented by a "violin" shape, which consists of a symmetrical kernel density plot mirrored along the y-axis. The width of each violin represents the density or proportion of data points at a given value. Wider sections indicate higher density, while narrower sections indicate lower density.

Inside the violin shape, typically centered, there is a box plot that displays additional information about the distribution of the data. The box plot usually includes a horizontal line for the median, a box indicating the interquartile range (IQR), and "whiskers" extending to the minimum and maximum values within a certain range or according to a specific criterion (e.g., within 1.5 times the IQR).

Violin plots can be particularly useful when comparing multiple groups and their distributions simultaneously. They allow for visualizing the overall shape of the data, including information about central tendency, spread, and multimodality, while also highlighting the relative density or frequency of values at different points.

```python
# Violin plots
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
sns.violinplot(x='species', y='sepal length (cm)', data=df)
plt.subplot(2, 2, 2)
sns.violinplot(x='species', y='sepal width (cm)', data=df)
plt.subplot(2, 2, 3)
sns.violinplot(x='species', y='petal length (cm)', data=df)
plt.subplot(2, 2, 4)
sns.violinplot(x='species', y='petal width (cm)', data=df)
plt.tight_layout()
plt.show()
```
