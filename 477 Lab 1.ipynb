```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 50 random numbers
np.random.seed(0)  # For reproducibility
random_numbers = np.random.rand(50)

# Define smoothing function using bin means method
def smooth_data(data, window_size=3):
    smoothed_data = np.zeros_like(data)
    for i in range(len(data)):
        start_index = max(0, i - (window_size - 1) // 2)
        end_index = min(len(data), i + (window_size - 1) // 2 + 1)
        smoothed_data[i] = np.mean(data[start_index:end_index])
    return smoothed_data

# Apply smoothing
smoothed_numbers = smooth_data(random_numbers)

# Plot the original and smoothed data
plt.figure(figsize=(10, 6))
plt.plot(random_numbers, label='Original Data', marker='o')
plt.plot(smoothed_numbers, label='Smoothed Data', linestyle='--')
plt.title('Smoothing by Bin Means Method')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
