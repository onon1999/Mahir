# 4. Write a Python program to calculate the difference between a given number and 17. If the number is greater than 17, return twice the absolute difference

```python
def difference(n):
    if n <= 17:
        return 17 - n
    else:
        return (n - 17) * 2

print(difference(19))
print(difference(14))
```

# 8. Write a Python program that computes the greatest common divisor (GCD) of two positive integers.

Sample Output:

GCD of 12 & 17 = 1

GCD of 4 & 6 = 2

GCD of 336 & 360 = 24

```python
def gcd(x, y):
   gcd = 1
   if x % y == 0:
       return y
   for k in range(int(y / 2), 0, -1):
       if x % k == 0 and y % k == 0:
           gcd = k
           break
   return gcd
print("GCD of 12 & 17 =",gcd(12, 17))
print("GCD of 4 & 6 =",gcd(4, 6))
print("GCD of 336 & 360 =",gcd(336, 360))
```

# 9. Write a Python program that removes and prints every third number from a list of numbers until the list is empty.

```python

def removeThirdNumber(int_list):


    pos = 3 - 1
    index = 0
    len_list = (len(int_list))


    while len_list > 0:

        index = (pos + index) % len_list


        print(int_list.pop(index))
        len_list -= 1

nums = [1, 2, 3, 4,5,6]
removeThirdNumber(nums)
```
